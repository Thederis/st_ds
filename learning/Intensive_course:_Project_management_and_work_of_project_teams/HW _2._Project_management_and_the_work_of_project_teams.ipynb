## Вопрос 1
На что больше похож подход CRISP-DM?
   - Agile;
   - Waterfall;
   - Эти подходы равнозначны.
   
   Обоснуйте свой ответ в развёрнутой форме.

## Ответ
Можно сравнить CRISP-DM с каскадной моделью (Waterfall), так как на этапе понимания бизнеса составляется подробный план проекта с описанием для каждого этапа — отличительный аспект традиционного каскадного подхода, который требует детального, предварительного планирования.

Если вы точно следуете CRISP-DM, определяя детальные планы для каждой фазы в начале проекта и включая каждый отчёт, и предпочитаете не проводить частые итерации, то вы работаете скорее по каскадному методу.

Однако, с другой стороны, CRISP-DM косвенно поддерживает принципы и практику Agile, где последовательность фаз не является жёсткой. Всегда требуется перемещение вперёд и назад между различными фазами. Результат каждой фазы определяет, какая фаза или конкретная задача фазы должна быть выполнена следующей.

Таким образом, если вы будете следовать CRISP-DM более гибко, быстро итерировать и включать в него другие Agile-процессы, то в итоге получите Agile-подход.

## Вопрос 2
В чём отличие инкрементной модели от итеративной?

## Ответ
Итеративная модель разработки предполагает разделение процесса разработки на несколько итераций. Каждая итерация включает в себя выполнение всех этапов жизненного цикла проекта, включая анализ, проектирование, реализацию и тестирование. После каждой итерации происходит обновление требований, доработка предыдущих этапов и/или добавление новой функциональности. Таким образом, итерации позволяют поэтапно уточнять и улучшать проект на основе обратной связи.

С другой стороны, инкрементная модель разработки предполагает последовательное создание и поэтапное добавление новой функциональности, формируя проект "кусочками" или инкрементами. Каждый новый инкремент представляет собой незавершенную версию проекта, которая, по мере добавления новых компонентов, становится все более полной и функциональной.

Таким образом, основное отличие между инкрементной и итеративной моделями заключается в том, что инкрементная модель сосредоточена на последовательном расширении функциональности проекта через добавление новых компонентов, в то время как итеративная модель ориентирована на поэтапное уточнение и развитие проекта на основе обратной связи после каждой итерации.

## Вопрос 3
Какая модель подойдёт для проекта по созданию нового Боинга и почему?

## Ответ
Боинг, как проект создания нового воздушного судна, является критически важным и высокосложным процессом, требующим строгого контроля над требованиями, четкого управления рисками и возможности внесения изменений на различных этапах разработки. С учетом этих особенностей, методология Spirai Model может быть подходящей для проекта по созданию нового Боинга по следующим причинам:

1. Итеративность и инкрементальность: Spirai Model, подразумевает пошаговое совершенствование и дополнение функциональности продукта на каждой итерации. Для создания такого сложного продукта, как новый Боинг, важно иметь возможность постепенно развивать и совершенствовать его функциональность.

2. Гибкость и адаптивность: запросы на изменение в требованиях на более позднем этапе могут быть точно включены с помощью этой модели.

3. Управление рисками: Spirai Model предусматривает регулярное проведение итераций, что способствует более эффективному выявлению и управлению рисками на различных этапах разработки продукта.

4. Подходит для больших проектов: рекомендуется использовать спиральную модель в больших и сложных проектах.

5. Обратная связь и уточнение требований: Методология Spirai Model стимулирует активное взаимодействие с заказчиком, что дает возможность непрерывно уточнять, обсуждать и корректировать требования на протяжении всего процесса разработки.

Таким образом, учитывая сложность и потребность в гибкости и контроле в разработке нового воздушного судна, методология Spirai Model может быть подходящей благодаря своим итеративным и инкрементальным принципам, гибкости, управлению рисками и акценту на постоянную обратную связь с заказчиком.

## Вопрос 4
Возьмите любой ИИ-проект, распишите для него план, декомпозируйте его и подберите подход. Объясните свой выбор.

## Ответ
Рассмотрим проект по созданию системы машинного обучения для предсказания спроса на продукцию в розничной торговле.

План разработки проекта с использованием CRISP-DM:
1. Изучение и анализ бизнес-требований
2. Изучение и анализ данных
3. Подготовка данных
4. Моделирование
5. Оценка
6. Внедрение

Декомпозиция проекта:
1. Изучение и анализ бизнес-требований:
   - Определение целей модели (предсказание спроса, оптимизация запасов)
   - Оценка ситуации (определение доступных ресурсов и требований к проекту, оценка рисков и непредвиденных обстоятельств)
   - Определение целей получения данных
   - Составление плана проекта
2. Изучение и анализ данных:
   - Сбор и подготовка данных по продажам, инвентарю, конкурентной активности и другим факторам
   - Описание данных(изучение данных и их свойств, таких как формат, количество записей и т.д.)
   - Исследование данных
   - Проверка качества данных (насколько данные чисты или грязны)
3. Подготовка данных:
   - Выбор данных (определение, какие наборы данных будут использоваться)
   - Очистка данных (исправление или удаление ошибочных значений)
   - Конструирование данных (выведение новых атрибуты, которые будут полезны)
   - Интеграция данных (создание новых наборов данных путём объединения из нескольких источников)
   - Форматирование данных
4. Моделирование:
   - Выбор подходящих моделей прогнозирования (временные ряды, регрессионный анализ, глубокое обучение)
   - Создание тестовой схемы (в зависимости от подхода к моделированию разделить данные на обучающую, тестовую и проверочную выборки)
   - Обучение и настройка моделей на имеющихся данных
   - Оценка модели
5. Оценка:
   - Оценка результатов (оценка, соответствуют ли модели критериям успеха бизнеса. Какая модель или модели должны быть одобрены для внедрения в бизнес)
   - Анализ процесса (всесторонний анализ проделанной работы)
   - Определение следующих шагов (на основании результатов предыдущих задач принять решение о том, следует ли приступить к развёртыванию, продолжить итерации моделей или начать новые проекты)
6. Внедрение:
   - Планирование развёртывания (разработка подробного плана развёртывания модели)
   - Обеспечение стабильной работы модели и ее мониторинг (комплексный план мониторинга и обслуживания модели для предотвращения любых проблем на этапе эксплуатации или после завершения проекта)
   - Подготовка заключительного отчёта
   - Обзор проекта (проведение ретроспективного анализа проекта, выявив успехи, области для улучшения и стратегии для будущего совершенствования.

Подход CRISP-DM (CRoss Industry Standard Process for Data Mining) выбирается из-за своей структурированной методологии, охватывающей все этапы жизненного цикла проекта от начала до конца. Он предоставляет гибкую, но структурированную рамку для работ по проектированию, разработке и внедрению системы прогнозирования спроса. Этот метод также уделяет внимание валидации модели, ее оценке и интеграции, что является важным аспектом успешного развертывания модели прогнозирования спроса в розничной торговле.
