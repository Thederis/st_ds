В этом разделе мы будем работать с со следующими библиотеками *Python*:

`numpy` — для матричных операций, статистических функций, функций линейной алгебры;

`sympy` — для работы с символьными выражениями, математическими символами и уравнениями;

`scipy` — для работы с массивами, линейной алгеброй, статистикой, оптимизацией;

`matplotlib` — для создания графиков и визуализации данных;

`pandas` — для работы с табличными данными;

`scikit-learn` — для реализации базовых алгоритмов машинного обучения.
