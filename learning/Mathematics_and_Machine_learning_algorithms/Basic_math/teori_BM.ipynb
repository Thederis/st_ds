## ОБЩИЕ ТЕРМИНЫ

**Линейная алгебра (linear algebra)** — это раздел математики, который изучает линейные пространства и линейные отображения между ними. Линейная алгебра — основа для многих методов машинного обучения, таких как линейная регрессия, метод главных компонент (PCA), SVM, нейронные сети и т. д.

**Математический анализ (calculus)** — это дисциплина, которая занимается изучением функций и их свойств. Математический анализ включает в себя изучение границ, непрерывности, дифференцируемости и интегрирования функций. В машинном обучении используется для оптимизации функций потерь, обучения моделей и других аспектов.

**Численные методы (numerical analysis)** — это набор методов для приближённого решения математических задач. Они используются в машинном обучении для решения задач оптимизации, аппроксимации функций и т. д.

**Теория вероятностей и математическая статистика (probability theory and mathematical statistics)** — это дисциплины, которые позволяют оценивать вероятности событий и делать выводы на основе данных. В машинном обучении они используются для создания моделей, оценки их точности, проверки гипотез и т. д.

**Искусственный интеллект (ИИ, artificial intelligence, AI)** — это область информатики, которая занимается разработкой алгоритмов и систем, способных выполнять задачи, которые обычно требуют умственных способностей человека. Примеры таких задач: распознавание речи и изображений, обработка естественного языка, принятие решений на основе данных и другие.

Термин «искусственный интеллект» включает в себя множество областей: машинное обучение, глубинное (глубокое) обучение и нейронные сети.

Существует две трактовки искусственного интеллекта: узкий искусственный интеллект и широкий искусственный интеллект. Они отличаются друг от друга по уровню способности к решению различных задач.

* **Узкий искусственный интеллект** способен решать только определённый набор задач, для которых он был создан. Эти задачи могут включать в себя распознавание речи, обработку естественного языка, игры, автоматизацию задач и т. д.
Примеры узкого ИИ — голосовые помощники Алиса и Siri, которые могут выполнять ограниченный набор задач: открытие приложений, поиск информации в интернете, установка таймера и т. д.

* В идеале **широкий искусственный интеллект** не имеет ограничений и может решать задачу из любой области. Однако сейчас не существует моделей, полностью соответствующих таким характеристикам. Пока что есть системы, в которых комплексно решается несколько задач узкого ИИ: например, модели, используемые в автономных автомобилях, роботах-хирургах и системах управления инфраструктурой. Но стоит отметить, что появляются большие модели, способные решать очень широкий спектр задач, например GPT-4.

В этом разделе мы изучим базовые модели узкого искусственного интеллекта.

**Машинное обучение (machine learning, ML)** — это подраздел искусственного интеллекта, который позволяет компьютерам учиться и делать предсказания на основе анализа больших объёмов данных. Машинное обучение использует алгоритмы и статистические модели, чтобы извлекать информацию из данных и делать предсказания или принимать решения.

Использование машинного обучения в различных областях (медицина, финансы, промышленность, розничная торговля и другие) позволяет существенно увеличить эффективность и точность принятия решений, а также автоматизировать рутинные задачи.

**Глубинное обучение (deep learning)** — это подраздел машинного обучения, который использует нейронные сети с большим количеством слоёв для анализа данных и выявления скрытых закономерностей.

Основная идея глубинного обучения — создание модели, которая может автоматически извлекать важные признаки из данных без необходимости ручного определения этих признаков. Для этого глубинные нейронные сети используют многослойную архитектуру, которая позволяет обрабатывать данные на разных уровнях абстракции.

В отличие от традиционных методов машинного обучения, где признаки данных должны быть явно определены и предварительно выбраны, глубинное обучение позволяет моделировать данные с большой выразительностью и гибкостью. Это делает его особенно полезным для решения сложных задач: например, для распознавания образов, обработки естественного языка, распознавания речи, анализа временных рядов и т. д.

**Наука о данных (data science, DS)** — это междисциплинарная область знаний, которая объединяет методы, инструменты и технологии для извлечения знаний и информации из данных.

*Data science* объединяет знания и методы из математики, статистики, информатики, машинного обучения, баз данных и визуализации данных. Это помогает анализировать и извлекать знания из больших, разнообразных, сложных и неструктурированных данных.

**Анализ данных (data analysis)** — это процесс извлечения полезной информации из данных для понимания явлений и процессов, лежащих в их основе, а также для принятия решений на основе полученных результатов.

Анализ данных включает в себя различные методы: статистический анализ, машинное обучение, исследование данных и другие. Цель анализа данных — выявление закономерностей, паттернов и трендов в данных, выделение важных факторов, влияющих на результаты, и прогнозирование будущих событий.

**Обучение с учителем (supervised learning)** — это метод машинного обучения, при котором модель обучается на размеченных данных, где каждый пример имеет метку класса или значение целевой переменной.

**Обучение без учителя (unsupervised learning)** — это метод машинного обучения, при котором модель обучается на неразмеченных данных, где нет меток классов или целевой переменной. Цель состоит в том, чтобы выявить структуру данных и паттерны.

**Обучение с подкреплением (reinforcement learning, RL)** — это метод машинного обучения, при котором модель обучается, взаимодействуя с окружающей средой и получая обратную связь в виде награды или штрафа.

## ОСНОВНЫЕ КОМПОНЕНТЫ

В машинном обучении есть множество специальных терминов, которые описывают различные концепции, методы и алгоритмы. Рассмотрим основные из них — начнём с компонентов.

**Объекты (objects)** — это наблюдаемые данные, которые необходимо обработать и проанализировать. Объекты могут представлять собой любые данные, которые можно описать числами или другими количественными характеристиками: звуковыми сигналами, изображениями, текстовыми документами, числовыми таблицами и т. д.

Каждый объект в машинном обучении представлен набором признаков или характеристик, которые описывают его свойства и состояние. Например, для изображения объектом могут быть пиксели, а для текстового документа — слова или символы.

В машинном обучении объекты используются для обучения моделей, которые могут классифицировать или регрессировать новые объекты на основе обучающих данных. Кроме того, объекты можно применять для оценки качества моделей и проведения исследовательского анализа данных.

**Признаки (features)** — это измерения или характеристики объекта, которые используются для его описания и анализа. Признаки могут быть числовыми, бинарными или категориальными в зависимости от типа данных, которые они описывают.

Например, если объектом является человек, то признаками могут быть его возраст, рост, вес, пол, образование, занятость и т. д. Для изображения признаками могут быть яркость и цвет каждого пикселя, а для текста — количество вхождений каждого слова.

Признаки играют важную роль в машинном обучении, поскольку на их основе строятся модели для решения задач классификации, регрессии, кластеризации и т. д. Выбор и определение правильных признаков — важный шаг в обработке и подготовке данных, так как неинформативные или некорректно выбранные признаки могут привести к плохим результатам моделирования.

**Целевая переменная (target variable)** — это переменная, которую модель пытается предсказать или прогнозировать на основе других переменных (признаков или характеристик). Целевая переменная может быть числовой или категориальной в зависимости от типа решаемой задачи.

* В задачах регрессии целевая переменная представляет собой непрерывную числовую переменную, которую модель пытается предсказать. Примеры задач регрессии: предсказание цены недвижимости или дохода клиента.

* В задачах классификации целевая переменная представляет собой категориальную переменную, которую модель пытается определить для новых данных. Примеры задач классификации: определение того, является электронное письмо спамом или не спамом, или определение класса заболевания на основе медицинских данных.

Выбор правильной целевой переменной критически важен в машинном обучении, так как это определяет тип задачи и алгоритмы для обучения модели. Также важно хорошо понимать, какие признаки могут быть наиболее полезными для предсказания целевой переменной.

**Набор данных, датасет (dataset)** — это коллекция данных, которые используются для обучения и тестирования моделей машинного обучения. Набор данных может содержать множество примеров, каждый из которых представляет собой некоторый объект или явление, и для них уже известны значения признаков (*features*) и целевых переменных (*target variables*) для предсказания.

Для построения модели машинного обучения требуется набор данных, который будет использоваться для обучения (*training* *dataset*) и для проверки качества модели (*validation dataset* или *test dataset*). Обычно набор данных разбивается на две части: одна используется для обучения модели, а другая — для её проверки. Размеры частей могут варьироваться, но обычно на обучение выделяется от 70 до 90 % данных, а на проверку — от 10 до 30 %.

Набор данных необходимо предварительно обработать и очистить от выбросов, пропусков и ошибок. Также может потребоваться преобразование данных для улучшения их качества или для адаптации под конкретный алгоритм машинного обучения.

Наборы данных могут быть различных типов: текстовые, изображения, звуковые, видео и другие. Также у них могут быть различные источники: измерения, эксперименты, интернет и т. д.

**Модель** — это математическое представление, которое используется для решения конкретной задачи, например для предсказания значения целевой переменной на основе данных признаков. Модель обучается на наборе данных, который включает в себя объекты, признаки и целевую переменную. После обучения её можно использовать на новых данных для предсказания значений целевой переменной.

Модель — это алгоритм, который можно реализовать на компьютере. Обучение модели заключается в такой настройке её параметров, чтобы ошибка прогноза на обучающих данных была минимальной. Процесс обучения может выглядеть по-разному в зависимости от типа задачи и выбранного алгоритма.

Типы моделей включают в себя линейные модели, деревья решений, нейронные сети, градиентный бустинг и многие другие. Каждая модель имеет свои сильные и слабые стороны, и её выбор зависит от требуемой точности, сложности задачи, объёма данных и других факторов.

Кроме того, одну и ту же модель можно использовать для разных задач, включая задачи классификации, регрессии, кластеризации и другие. Важно правильно выбрать модель и метод её обучения для конкретной задачи, чтобы добиться наилучшей точности и эффективности.

**Функция потерь (loss function)** — это математическая функция, которая оценивает разницу между предсказанными значениями модели и фактическими значениями на обучающем наборе данных. Цель функции потерь — минимизировать разницу между предсказанными и фактическими значениями, чтобы модель могла давать наиболее точные прогнозы на новых данных.

* В задачах регрессии, когда требуется предсказать численное значение (например, цену дома), в качестве функции потерь можно выбрать среднеквадратичную ошибку (*MSE*) или среднюю абсолютную ошибку (*MAE*). 

* В задачах классификации, когда требуется предсказать категорию или метку, в качестве функции потерь можно взять перекрёстную энтропию (*cross-entropy*) или логистическую функцию потерь.

В машинном обучении выбор функции потерь зависит от задачи, типа модели и используемого алгоритма обучения. Если верно её выбрать, возрастёт точность модели и скорость её обучения. В некоторых случаях может понадобиться более сложная функция потерь, которая учитывает не только разницу между предсказанными и фактическими значениями, но и другие факторы, например сложность модели или балансировку классов в задаче классификации.

Математическая нотация приведённых выше определений:

 $x_i$ — объект, для которого делается предсказание;

 $d$ — размерность;

 $X$ — пространство объектов;

 $x_{i1},x_{i2},...,x_{id}$ — набор из $d$ признаков для $i$-го объекта;

 $y_i$ — ответ для объекта $i$, целевая переменная;

 $Y$ — пространство ответов;

 $n$ — количество примеров, размер обучающей выборки;

 $X = (x_i,y_i)^n_{i=1}$ — обучающая выборка;

 $m(x_i)$-модель — функция, предсказывающая ответ для объекта $x_i$ (можно сказать, что $m$ отображает пространство $X$ в пространство $Y$);

 $\hat y_i$ — предсказание модели для объекта $i$;

 $L(y_i,\hat y_i)$ — функция потерь, показатель того, насколько предсказание модели отличается от реального ответа на конкретном примере.

## ОСНОВНЫЕ ТИПЫ ЗАДАЧ

Давайте разберёмся в основных типах задач в машинном обучении.

**Регрессия (regression)** — это метод для анализа отношений между целевой (зависимой) переменной и одной или несколькими независимыми переменными (признаками). Регрессия часто используется для прогнозирования количественных значений: цена, объём продаж, количество посетителей и т. д.

В задаче регрессии целевая переменная является количественной, то есть может принимать любые числовые значения. Цель регрессионного анализа — построить модель, которая может предсказывать значения целевой переменной на основе значений признаков.

Алгоритм для построения модели регрессии находит такие значения параметров, которые минимизируют ошибку прогнозирования на обучающих данных. Алгоритмов регрессии много: линейная регрессия, полиномиальная регрессия, деревья решений, случайный лес и другие.

Одна из важных задач в регрессии — оценка качества модели на новых данных. Для этого используется тестовый набор данных. Чтобы оценить качество модели, применяют различные метрики: среднеквадратичную ошибку (*MSE*), среднюю абсолютную ошибку (*MAE*) и другие.

**Классификация (classification)** — это процесс отнесения объектов к заранее определённым классам или категориям. Классификация используется для широкого спектра задач: распознавание образов, определение темы текста, диагностика заболеваний, прогнозирование рынка и мн. др.

Алгоритмы в задачах классификации обучаются на наборе данных, состоящем из объектов и соответствующих им меток классов (целевых переменных). Обычно для обучения модели используется алгоритм, который находит зависимость между признаками объектов и их метками классов, а затем использует эту зависимость для прогнозирования меток классов для новых объектов.

Примеры алгоритмов классификации: деревья решений, наивный байесовский классификатор, метод опорных векторов (*SVM*), логистическая регрессия, градиентный бустинг и нейронные сети. В зависимости от природы задачи и характеристик данных один алгоритм может быть более эффективным, чем другой.

Для оценки качества работы алгоритма классификации используются различные метрики: доля верно предсказанных ответов (*accuracy*), полнота (*recall*), точность предсказания (*precision*), *F1*-мера и другие. Для настройки параметров модели и выбора наилучшего алгоритма классификации используется кросс-валидация на обучающих данных.

**Кластеризация (clusterization)** — это задача разбиения множества объектов на группы (кластеры) таким образом, чтобы объекты внутри каждого кластера были максимально похожи между собой, а объекты из разных кластеров были максимально различными.

Кластеризация используется во многих областях: маркетинг, медицина, социология, биология и другие. Например, в маркетинге кластеризацию можно использовать для выявления групп потребителей со схожими предпочтениями и поведением, а в медицине — для разбиения пациентов на группы на основе сходства симптомов и лечения.

Для решения задачи кластеризации используются различные алгоритмы: K-средних (*K-means*), *DBSCAN*, иерархическая кластеризация и т. д. Алгоритмы кластеризации могут работать с разными типами данных, включая числовые, категориальные и текстовые.

Оценить качество кластеризации может быть сложно, так как нет точного определения того, как должны быть сформированы кластеры. Для оценки качества кластеризации используются различные метрики: коэффициент силуэта, индекс Данна и др.

**Снижение размерности (dimensionality reduction)** — это процесс уменьшения количества признаков в наборе данных. При этом сохраняется максимально возможный объём информации о данных. Это позволяет упростить анализ данных и ускорить вычисления в алгоритмах машинного обучения.

Снижение размерности может служить как для уменьшения количества шума и выбросов в данных, так и для упрощения модели при сохранении информативных признаков.

Снижение размерности может пригодиться во многих задачах машинного обучения: классификации, кластеризации, визуализации данных и других.

Помимо приведённых выше типов задач, в машинном обучении встречаются и другие: обучение с подкреплением, детектирование аномалий и прочие. Их изучение выходит за рамки тем этого раздела.

## ЧТО ТАКОЕ ВЕКТОР?

**Вектор** — это математический объект, который имеет длину и направление в пространстве. С помощью векторов можно представить физические величины — силу, скорость, ускорение и др.

В машинном обучении вектор — одно из основных понятий для описания данных. Например, вектор может представлять цвет изображения в *RGB*-формате, размер объекта на изображении, частоту звука или другие характеристики данных.

Векторы также используются в алгоритмах машинного обучения для выполнения классификации, кластеризации и регрессии. Кроме того, векторы можно использовать для решения задач оптимизации.

Каждый объект может быть представлен в виде вектора признаков, где каждый элемент вектора соответствует определённому признаку объекта.

Векторы обрабатывают с помощью различных методов машинного обучения: линейной регрессии, метода опорных векторов, нейронных сетей, деревьев решений и др.


В *Python* вектор чаще всего бывает представлен в виде одномерного *numpy*-массива.

Векторы обычно обозначаются с помощью **полужирного** шрифта или стрелки над символом: $u$ или $\vec u$. В общем случае в $n$-мерном пространстве вектор представляется как набор из $n$ чисел.

Компоненты вектора могут быть действительными или комплексными числами. Для записи компонент вектора обычно используют формат столбца (вектор-столбец), где каждый элемент записывается в новой строке:

$$\mathbf{v} = \vec v = \left[\begin{matrix}v_1\cr v_2\cr \vdots\cr v_n \end{matrix}\right]$$

Элементы вектора $v_1,v_2,v_3, ...,v_{n−1},v_n$ называются его **координатами**, или **компонентами**. **Размерность** вектора — это количество его компонент.

Существует связь между понятием вектора в -мерном пространстве и упорядоченной последовательностью действительных чисел $v_1,v_2,v_3, ...,v_{n−1},v_n$.

В двухмерном пространстве вектор можно представить как пару чисел $(v_1,v_2)$, где $v_1,v_2$ можно интерпретировать как координаты $x,y$ на плоскости. В трёхмерном пространстве векторы задаются с помощью трёх координат. Векторы большей размерности нельзя визуализировать.

Некоторые типы векторов имеют собственные названия, например **нулевой вектор** — это вектор, все компоненты которого равны нулю:
$$\overrightarrow{0}=\left[\begin{array}{c} 0 \\ 0 \\ \vdots \\ 0 \end{array}\right]$$

В машинном обучении принято задавать векторы с помощью только одной точки в заданном векторном пространстве. Соответственно, предполагается, что **начало всех векторов находится в точке начала координат**. Такая интерпретация векторов делает интуитивно понятными различные операции над векторами, например вычисление угла между двумя векторами:

### ОПРЕДЕЛЕНИЕ НАПРАВЛЕНИЯ ВЕКТОРА

Направление вектора определяется от его начала к его концу. Другими словами, **направление вектора** — это угол, под которым он направлен относительно некоторой начальной точки (обычно это начало координат) в пространстве.

Обычно этот угол измеряется в градусах или радианах. Например, для вектора с координатами (5,4) угол между горизонтальной осью и вектором можно найти, используя тангенс угла наклона, который равен $\frac{4}{5}$.

Чтобы определить угол по тангенсу, можно воспользоваться обратной функцией тангенса — арктангенсом (или тангенсом в степени -1). Арктангенс обозначается как $\text{arctan}$ или $\tan^{-1}$. Если дано значение тангенса угла , то угол  можно найти следующим образом:
$$\alpha=\text{arctan}(\tan{\alpha})$$

Например, если дано, что $\text{arctan}$,то угол $\alpha$ можно найти следующим образом:
$$\alpha=\text{arctan}(0.8)\approx 38.659^{\circ}$$
Однако необходимо помнить, что арктангенс имеет ограничения на область определения и может возвращать только углы в определённом диапазоне значений. Например, для обычной арктангенс-функции область определения ограничена значениями $-\frac{\pi}{2} < \text{arctan}(x)<\frac{\pi}{2}$. То есть функция возвращает только углы, лежащие в этом диапазоне.

Если необходимо получить угол, лежащий в другом диапазоне, можно использовать дополнительные математические операции. Например, можно добавлять или вычитать $\pi$ (180 градусов) из значения арктангенса в зависимости от того, в какой четверти находится искомый угол.

Соответственно, направление вектора с координатами (5,4) равно $\approx 38.659^{\circ}$.

### ВЫЧИСЛЕНИЕ ДЛИНЫ ВЕКТОРА

Каждый вектор (кроме нулевого) в векторном пространстве имеет длину. Длина вектора обозначается термином «**норма**».

Существуют несколько способов определения нормы векторов — все они моделируют расстояния, которые мы измеряем в повседневной жизни.

**L1-НОРМА**

**L1-норма** (расстояние городских кварталов, манхэттенская метрика) соответствует расстоянию, которое проехал бы автомобиль между начальной и конечной точками по городским улицам, расположенным под углом $90^{\circ}$. Это расстояние находится путём перемещения только вдоль осей координат:
$$\|\vec{v}\|_1=\left(\sum_{i=1}^d\left|v_i\right|^1\right)^{\frac{1}{1}}$$
Здесь $v$ — вектор размерности $d$.

В этой формуле суммируются модули всех координат. Корень первой степени означает отсутствие математических действий над выражением и используется лишь для унификации с метрикой Минковского.

К примеру, расстояние от точки (0,0) до точки (5,4) может быть пройдено по разным траекториям. Однако общее расстояние останется равным 9.

**L2-НОРМА**

**L2-норма** соответствует расстоянию, которое пролетел бы вертолёт между началом координат и точкой вектора. Это расстояние находится по теореме Пифагора: сначала суммируются квадраты всех координат, а затем вычисляется корень второй степени из этой суммы:
$$\|\vec{v}\|_2=\left(\sum_{i=1}^d\left|v_i\right|^2\right)^{\frac{1}{2}}$$
Здесь $v$ — вектор размерности $b$.

*L2*-норма более популярна по сравнению с *L1*-нормой, потому что она более естественна для измерения длины:

**МЕТРИКА МИНКОВСКОГО**

Общая формула нормы вектора в $\mathbb{R}^d$ носит название метрики Минковского и вычисляется по формуле:
$$\|\vec{v}\|_n=\left(\sum_{i=1}^d\left|v_i\right|^n\right)^{\frac{1}{n}}$$
Здесь $v$ — вектор размерности $d,n$ — показатель степени.

Запись $\mathbb{R}^d$ означает, что мы рассматриваем векторы, элементы которых являются числами из множества действительных чисел $\mathbb{R}$, и каждый вектор имеет $d$ компонент (размерность $d$). Таким образом, векторы в $\mathbb{R}^d$ можно записать в виде упорядоченных наборов чисел вида $(x_1,x_2, ...,x_d)$, где каждое $x_i$ принадлежит множеству действительных чисел $\mathbb{R}$.

Примеры:

Вектор (2, 5, -3) принадлежит векторному пространству $\mathbb{R}^3$.
Вектор (1, 0, 1, -2) принадлежит векторному пространству $\mathbb{R}^4$.

### СЛОЖЕНИЕ И ВЫЧИТАНИЕ ВЕКТОРОВ

Векторы, как и числа, можно складывать:
$$\overrightarrow{v 1}+\overrightarrow{v 2}=\left[\begin{array}{c} v 1_1 \\ v 1_2 \\ \vdots \\ v 1_n \end{array}\right]+\left[\begin{array}{c} v 2_1 \\ v 2_2 \\ \vdots \\ v 2_n \end{array}\right]=\left[\begin{array}{c} v 1_1+v 2_1 \\ v 1_2+v 2_2 \\ \vdots \\ v 1_n+v 2_n \end{array}\right]$$
и вычитать друг из друга, получая новые векторы:
$$\vec{v}+(-v \overrightarrow{2})=\left[\begin{array}{c} v 1_1 \\ v 1_2 \\ \vdots \\ v 1_n \end{array}\right]+\left[\begin{array}{c} -v 2_1 \\ -v 2_2 \\ \vdots \\ -v 2_n \end{array}\right]=\left[\begin{array}{c} v 1_1-v 2_1 \\ v 1_2-v 2_2 \\ \vdots \\ v 1_n-v 2_n \end{array}\right]$$
Например, чтобы сложить вектор $\vec{v1}$ с координатами (4,7) (на рисунке (1) — красного цвета) и вектор $\vec{v2}$ с координатами (8,4) (на рисунке (1) — синего цвета), нужно просто сложить их соответствующие координаты, и тогда получится новый вектор $\vec{v1+v2}$ с координатами (12,11) (на рисунке (1) — зелёного цвета).

Геометрически это можно представить так: суммарный вектор является диагональю параллелограмма, образованного векторами.

Вычитание происходит аналогично, только используется другая диагональ параллелограмма. Например, разность между векторами $\vec{v1}$ (4,7) (на рисунке (2) — красного цвета) и $\vec{v2}$ (8,4) (на рисунке (2) — синего цвета) вычисляется также поэлементно. Получается вектор $\vec{v1-v2}$ с координатами (-4,3) (на рисунке (2) — зелёного цвета):

### УМНОЖЕНИЕ ВЕКТОРА НА КОНСТАНТУ

Умножение вектора на константу — это поэлементное умножение компонент вектора на данную константу:
$$c \vec{v}=c\left[\begin{array}{c} v_1 \\ v_2 \\ \vdots \\ v_n \end{array}\right]=\left[\begin{array}{c} c v_1 \\ c v_2 \\ \vdots \\ c v_n \end{array}\right]$$
Умножение вектора на константу приводит к изменению длины вектора и/или его направления в зависимости от значения константы. Геометрический смысл умножения вектора на константу следующий:





<center><em>(1) — константа положительная, умножение приводит к увеличению длины вектора, направление не меняется</em></center>
<center><em>(2) — константа отрицательная, длина вектора также увеличивается, но направление меняется на противоположное</em>

**СВОЙСТВА УМНОЖЕНИЯ ВЕКТОРА НА КОНСТАНТУ**

Для любого вектора выполняется:

 * $c \cdot \vec{v}=\vec{v} \cdot c$ (коммутативность умножения на константу).

 * $(c_1\cdot c_2) \cdot \vec{v}=c_1 \cdot (c_2 \cdot \vec{v})$ (ассоциативность умножения на несколько констант).

### ТРАНСПОНИРОВАНИЕ ВЕКТОРА

**Транспонирование вектора** — это операция, при которой вектор записывается в виде матрицы размера $1 \times n$ (в случае вектора-столбца) или $n \times 1$ (в случае вектора-строки), где $n$ — количество элементов вектора.

При транспонировании вектора-столбца получается вектор-строка:
$$\vec{v}=[v_1 \ v_2 \ ... \ v_n]$$

### СКАЛЯРНОЕ ПРОИЗВЕДЕНИЕ ВЕКТОРОВ

**Скалярное произведение векторов** — это операция, результатом которой является скалярное значение (число), полученное путём умножения соответствующих элементов двух векторов и их последующей суммы.

Скалярное произведение можно вычислить, только если длины двух векторов равны.

Для скалярного произведения векторов $\vec{v1}$ и $\vec{v2}$ используется одно из следующих обозначений: $⟨\vec{v1},\vec{v2}$⟩, $⟨\vec{v1},\vec{v2}⟩$, $v1 \cdot v2$, $\vec{v1} \cdot \vec{v2}$.

Для двух векторов  и  длиной  скалярное произведение можно вычислить по формуле:

$⟨\vec{v1},\vec{v2}⟩ =\sum_{i=1}^{n}v1_iv2_i$, где $\vec{v1}=[\vec{v1}_1,\vec{v1}_2,...,\vec{v1}_n]$ и $\vec{v2}=[\vec{v2}_1,\vec{v2}_2,...,\vec{v2}_n]$ — два вектора длины $n$.

Скалярное произведение также можно записать в виде произведения вектора-строки и вектора-столбца.

**СВОЙСТВА СКАЛЯРНОГО ПРОИЗВЕДЕНИЯ ВЕКТОРОВ:**

* С помощью скалярного произведения можно вычислить длину вектора $\vec{v}$ как корень из скалярного произведения вектора на себя: $L_2-\text{norm}=\sqrt{\langle v,v\rangle}$.

* С помощью скалярного произведения можно вычислить угол между двумя векторами $\vec{v1}$ и $\vec{v2}$:
$\cos \theta=\frac{\langle\overrightarrow{v 1} \cdot \overrightarrow{v 2}\rangle}{|\overrightarrow{v 1}||\overrightarrow{v 2}|}$, где $\theta$ — угол между векторами, $|\vec{v1}|$ и $|\vec{v2}|$ — длины векторов $\vec{v1}$ и $$\vec{v1}$$ соответственно.

* С помощью скалярного произведения можно вычислить проекцию вектора $\vec{v2}$ на вектор $\vec{v1}$: $\operatorname{proj}_{\vec{v}_1} \vec{v}_2=\frac{\langle\vec{v1} \cdot \overrightarrow{v 2}\rangle}{\left|\vec{v1}\right|^2} \vec{v}_1$.



<center><em>Проекция вектора $v_2$ на вектор $v_1$</em>

* Для всех векторов, угол между которыми меньше $90^{\circ}$, скалярное произведение будет положительным.
Для ортогональных векторов (угол между ними равен $90^{\circ}$) скалярное произведение равно 0.

Для всех векторов, угол между которыми больше $90^{\circ}$, скалярное произведение будет отрицательным.



<center><em>Знак скалярного произведения при разных углах между векторами</em>

У скалярного произведения векторов множество применений в математике, физике и инженерии. Вот некоторые из них:

* **Решение систем линейных уравнений.** Систему линейных уравнений можно записать в виде матричного уравнения $Ax=b$, где $A$ — матрица коэффициентов, $x$ — вектор неизвестных, $b$ — вектор правых частей. Решить эту систему можно с помощью методов, основанных на скалярном произведении, например методом Гаусса.
* **Кластерный анализ.** В кластерном анализе скалярное произведение можно использовать для измерения сходства между объектами или для оценки близости кластеров.

### ВЕКТОРНОЕ ПРОИЗВЕДЕНИЕ ВЕКТОРОВ

**Векторное произведение** — это бинарная операция над двумя векторами, результатом которой является новый вектор, перпендикулярный обоим исходным векторам.

с координатами (1, 2, 0) (синий). Результатом является вектор $\vec{v_1 \times v_2}$ с координатами (0, 0, 6) (зелёный).

Компоненты вектора, полученного в результате векторного произведения, вычисляются как разности попарных произведений компонент исходных векторов, умноженных на коэффициенты $\pm 1$ в соответствии с правилом буравчика. Векторное произведение двух векторов $\vec{v_1}$ и $\vec{v_2}$ в трёхмерном пространстве вычисляется по формуле:
$$\vec{a} \times \vec{b}=\left(a_1, a_2, a_3\right) \times\left(b_1,  b_2, b_3\right)=\left(a_2 b_3-a_3 b_2 3 b_1-a_1 b_3 1 b_2-a_2 b_1\right)$$


***Правило буравчика (винта)** для векторного произведения: «Если отобразить векторы так, чтобы их начала совпадали, и вращать первый вектор-сомножитель кратчайшим образом ко второму вектору-сомножителю, то буравчик (винт), вращающийся таким же образом, будет завинчиваться в направлении вектора-произведения».*

Векторное произведение довольно редко используют в машинном обучении, однако оно может быть полезным для некоторых задач, например для создания новых признаков из двух или более существующих признаков.

### ВЫЧИСЛЕНИЕ РАССТОЯНИЯ МЕЖДУ ВЕКТОРАМИ

**ВЫЧИСЛЕНИЕ РАССТОЯНИЯ МЕЖДУ ВЕКТОРАМИ**

В машинном обучении часто приходится находить расстояние между двумя векторами. Для обозначения функции, измеряющей это расстояние, используется термин «метрика». Она должна удовлетворять следующим требованиям:

* **Неотрицательность**. Метрика должна быть неотрицательной, то есть расстояние между любыми двумя векторами не должно быть меньше нуля. Формально: d(x,y)\geq 0 для любых векторов $x$ и $y$, и $d(x,y)= 0$ тогда и только тогда, когда $x=y$.

* **Симметричность**. Метрика должна быть симметричной, то есть расстояние между векторами $x$ и $y$ должно быть таким же, как расстояние между $x$ и $y$. Формально: $d(x,y)= d(y,x)$ для любых векторов $x$ и $y$.

* **Неравенство треугольника**. Метрика должна удовлетворять неравенству треугольника, то есть расстояние между любыми двумя векторами через третий вектор должно быть меньше, чем сумма расстояний между первым и третьим и между вторым и третьим векторами. Формально: $d(x,y)\leq d(x,z)+ d(z,y)$ для любых векторов $x$, $y$ и $z$.

* **Тождественность**. Метрика должна удовлетворять тождеству: $d(x,y)= 0$ тогда и только тогда, когда $x=y$.

Метрика, удовлетворяющая этим требованиям, называется **метрикой расстояния.**

Для вычисления расстояния между векторами можно использовать те же способы, что и для определения длины вектора.

**L1-РАССТОЯНИЕ**

В случае двух векторов *L1*-расстояние вычисляется как расстояние между концами этих векторов. Строго говоря, такое расстояние называется суммой абсолютных значений разности компонент. Оно вычисляется по формуле:
$$|\overrightarrow{v 1}-\overrightarrow{v 2}|_1=\sum_{i=1}^n\left|v 1_i-v 2_i\right|$$
Здесь:

* $\vec{v1}$ и $\vec{v2}$ — два вектора размерности $n$.
* $v1$ и $v2$ — их соответствующие компоненты. 

**L2-РАССТОЯНИЕ**

Другой способ — *L2*-метрика, или *L2*-расстояние — вычисляется как квадратный корень из суммы квадратов разностей компонент по формуле:
$$|\overrightarrow{v 1}-\overrightarrow{v 2}|_2=\sqrt{\sum_{i=1}^n\left(v 1_i-v 2_i\right)^2}$$
Здесь:

* $\vec{v1}$ и $\vec{v2}$ — два вектора размерности $n$.  
* $v1$ и $v2$ — их соответствующие компоненты.

**КОСИНУСНОЕ РАССТОЯНИЕ**

**Косинусное расстояние**,или косинусная метрика — ещё один способ определения различий между векторами. Оно вычисляется как единица минус косинус угла между векторами по формуле:
$$\text { cosine distance }(\overrightarrow{v 1}, \overrightarrow{v 2})=1-(\overrightarrow{v 1} \cdot \overrightarrow{v 2}) /(|\overrightarrow{v 1}| *|\overrightarrow{v 2}|)$$
Здесь:

* $\vec{v1}$ и $\vec{v2}$ — скалярное произведение векторов $v1$ и $v2$.
* $|\vec{v1}|$ и $|\vec{v2}|$ — длины векторов $v1$ и $v2$.

В машинном обучении очень полезно знать меры расстояния между векторами, чтобы определять сходство между точками в данных.

## ЧТО ТАКОЕ МАТРИЦА?

**Матрица** — это прямоугольная таблица чисел, символов или выражений, разделённых на строки и столбцы. Матрицу можно представить как набор векторов-строк или векторов-столбцов, каждый из которых содержит элементы матрицы.

Обычно матрицу обозначают заглавными буквами латинского алфавита — например, $A$, $B$ или $C$. Размеры матрицы называются её размерностью и указывается в виде $m \times n$, где $m$ — количество строк, а $n$ — количество столбцов.

Формальная запись для матрицы:
$$A^{m \times n}=\left(a_{i j}\right) \in R^{m \times n}$$
Каждый элемент матрицы можно обозначить индексами ($i,j$), где $i$ — номер строки, а $j$ — номер столбца. Например, элемент $A_{i,j}$ означает элемент матрицы $A$ в $i$-ой строке и $j$-ом столбце.

Обычно для обозначения матриц используют квадратные скобки, но также возможны другие скобки и вертикальные линии:
$$A=\left[\begin{array}{ll} a_{11} & a_{12} \\ a_{21} & a_{22} \\ a_{31} & a_{32} \end{array}\right]=\left(\begin{array}{ll} a_{11} & a_{12} \\ a_{21} & a_{22} \\ a_{31} & a_{32} \end{array}\right)=\left\{\begin{array}{ll} a_{11} & a_{12} \\ a_{21} & a_{22} \\ a_{31} & a_{32} \end{array}\right\}=\left|\begin{array}{ll} a_{11} & a_{12} \\ a_{21} & a_{22} \\ a_{31} & a_{32} \end{array}\right|=\left\|\begin{array}{ll} a_{11} & a_{12} \\ a_{21} & a_{22} \\ a_{31} & a_{32} \end{array}\right\|$$
В машинном обучении матрицы используют для представления наборов данных, где каждая строка — это отдельный пример или объект, а каждый столбец — отдельный признак или характеристика.

## ОПЕРАЦИИ НАД МАТРИЦАМИ



### СЛОЖЕНИЕ И ВЫЧИТАНИЕ МАТРИЦ

Чтобы сложить или вычесть две матрицы $A$ и $B$, они должны иметь одинаковое количество строк и столбцов. Если бы мы имели дело с таблицами, то сказали бы, что все ячейки попарно складываются друг с другом.

Строго говоря, к элементу матрицы $A$ прибавляется или вычитается соответствующий элемент матрицы $B$:

$$A \pm B=\left[\begin{array}{ccc} a_{11} & \cdots & a_{1 n} \\ a_{21} & \cdots & a_{2 n} \\ \vdots & \ddots & \vdots \\ a_{m 1} & \cdots & a_{m n} \end{array}\right] \pm\left[\begin{array}{ccc} b_{11} & \cdots & b_{1 n} \\ b_{21} & \cdots & b_{2 n} \\ \vdots & \ddots & \vdots \\ b_{m 1} & \cdots & b_{m n} \end{array}\right]=\left[\begin{array}{ccc} a_{11} \pm \mathrm{b}_{11} & \cdots & a_{1 n} \pm \mathrm{b}_{1 n} \\ a_{21} \pm \mathrm{b}_{21} & \cdots & a_{2 n} \pm \mathrm{b}_{2 n} \\ \vdots & \ddots & \vdots \\ a_{m 1} \pm \mathrm{b}_{m 1} & \cdots & a_{m n} \pm \mathrm{b}_{m n} \end{array}\right]$$

### УМНОЖЕНИЕ МАТРИЦЫ НА КОНСТАНТУ

Чтобы умножить матрицу $A$ на число $c$, необходимо умножить каждый элемент матрицы $A$ на число $c$. Результатом будет новая матрица.
$$c \cdot A=c \cdot\left[\begin{array}{cccc} a_{11} & a_{12} & \cdots & a_{1 n} \\ a_{21} & a_{22} & \cdots & a_{2 n} \\ \vdots & \vdots & \ddots & \vdots \\ a_{m 1} & a_{m 2} & \cdots & a_{m n} \end{array}\right]=\left[\begin{array}{cccc} c \cdot a_{11} & c \cdot a_{12} & \cdots & c \cdot a_{1 n} \\ c \cdot a_{21} & c \cdot a_{22} & \cdots & c \cdot a_{2 n} \\ \vdots & \vdots & \ddots & \vdots \\ c \cdot a_{m 1} & c \cdot a_{m 2} & \cdots & c \cdot a_{m n} \end{array}\right]$$

### ТРАНСПОНИРОВАНИЕ МАТРИЦЫ

**Транспонирование матрицы** — это операция, при которой строки матрицы становятся её столбцами, а столбцы — строками. То есть матрица отражается относительно её главной диагонали.

Обозначение для транспонированной матрицы — символ $\top$ в верхнем правом углу:
$$A=\left[\begin{array}{lll} 1 & 2 & 3 \\ 4 & 5 & 6 \end{array}\right] \quad A^{\top}=\left[\begin{array}{ll} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{array}\right]$$

### ПРОИЗВЕДЕНИЕ МАТРИЦ

Умножение матрицы на матрицу — одна из основных операций линейной алгебры. Умножать матрицы можно только в том случае, когда вторая размерность первой матрицы равна первой размерности второй матрицы. Для того чтобы умножить матрицу $A$ размерности $m \times n$ на матрицу $B$ размерности $n \times p$, необходимо выполнить следующие действия:

Создать новую матрицу $C$. Если бы мы говорили про таблицы, то сказали бы, что новая таблица берёт меньшее количество столбцов и строк от исходной таблицы.

Строго говоря, мы получаем матрицу $C$ размерности $m \times p$, которая будет содержать результат умножения матриц $A$ и $B$.

Для каждого элемента матрицы $C_{i,j}$ вычислить его значение как сумму произведений элементов $i$-й строки матрицы #A# и соответствующих элементов -го столбца матрицы $B$:
$$C_{i, j}=A_{i, 1} * B_{1, j}+A_{i, 2} * B_{2, j}+\ldots+A_{i, n} * B_{n, j}$$

### ИСПОЛЬЗОВАНИЕ ЕДИНИЧНОЙ И ОБРАТНОЙ МАТРИЦ

**Единичная матрица** — это квадратная матрица, у которой на главной диагонали стоят единицы, а остальные элементы равны нулю.

Обычно единичную матрицу обозначают буквой $I$:
$$I=\left[\begin{array}{lll} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{array}\right] \quad \text { или } \quad I_{3 \times 3}$$

**Обратная матрица** — это матрица, умножение которой на исходную матрицу даёт единичную матрицу.

Обратную матрицу можно обозначить с помощью символа $-1$ в верхнем правом углу матрицы:
$$A=\left[\begin{array}{ccc} 2 & 5 & 7 \\ 6 & 3 & 4 \\ 5 & -2 & -3 \end{array}\right] \quad A^{-1}=\left[\begin{array}{ccc} 1 & -1 & 1 \\ -38 & 41 & -34 \\ 27 & -29 & 24 \end{array}\right] \quad I=\left[\begin{array}{ccc} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{array}\right]$$

Единичная и обратная матрицы активно используются в линейной алгебре — фундаменте многих методов машинного обучения.

* Единичную матрицу применяют при решении систем линейных уравнений. Если векторы-столбцы признаков $X$ имеют линейно независимые компоненты, то систему уравнений можно решить с помощью матрицы $X$ и единичной матрицы, которые связываются с помощью метода [Гаусса — Жордана](https://amkbook.net/mathbook/solving-slae-gauss-jordan-method "AMKBook.Net - Метод Гаусса-Жордана").

* Обратная матрица используется для решения линейных уравнений, задач оптимизации и настройки параметров моделей машинного обучения. Однако стоит отметить, что обратная матрица не всегда определена. Например, если определитель (ниже мы обсудим, что это) матрицы равен нулю, то обратная матрица не существует. Такое может произойти, когда количество признаков превышает количество наблюдений в выборке. Поэтому важно тщательно проверять матрицы на обратимость и, если возможно, использовать методы, которые не требуют обращения матриц.

Матрица, содержащая линейно зависимые строки, называется **сингулярной**, или **вырожденной**. Матрица, не содержащая линейно зависимых строк, называется **невырожденной**, или **обратимой**.*

Линейная зависимость признаков в машинном обучении порождает несколько проблем:

* **Переопределение модели.** Если признаки линейно зависимы, то можно построить множество моделей, которые дадут одинаковый результат. Это может привести к тому, что модель будет слишком сложной и переобученной. То есть она будет хорошо работать на обучающих данных, но плохо — на новых данных.

* **Низкая устойчивость.** Если признаки линейно зависимы, то модель может быть чувствительна к небольшим изменениям входных данных. Например, если один признак можно выразить через другой с небольшой ошибкой, то небольшое изменение входных данных может привести к значительным изменениям в результатах.

* **Плохая интерпретируемость.** Если признаки линейно зависимы, то сложно понять, какой признак вносит больший вклад в результаты модели. Из-за этого может быть проблематично анализировать результаты и принимать на их основе решения.

Мультиколлинеарность может привести к неправильной оценке коэффициентов модели и снижению точности прогнозирования. Кроме того, она может сделать модель менее интерпретируемой, что затрудняет понимание вклада каждой переменной в модель.

Если обнаружилась мультиколлинеарность, можно:

* исключить одну из коррелирующих переменных из модели;

* объединить коррелирующие переменные в одну переменную;

* использовать регуляризацию для уменьшения влияния некоторых переменных на модель.

Для выявления мультиколлинеарности можно использовать статистические методы: корреляционный анализ или метод главных компонент. Далее мы рассмотрим такие методы, как вычисление **ранга** и **определителя матрицы**.

*Если матрица признаков имеет низкий ранг или близкий к нулю определитель, это может указывать на наличие мультиколлинеарности в данных.*

### РАНГ МАТРИЦЫ

**Ранг матрицы** — это количество линейно независимых строк или столбцов в матрице.

Если матрица имеет ранг, равный меньшему из числа строк или столбцов, то она называется **вырожденной**.

Высокий ранг матрицы признаков означает, что в данных мало корреляции между признаками. Это может быть признаком отсутствия мультиколлинеарности.

Одна из интерпретаций ранга матрицы — **мера неизбыточной информации**, которая хранится в матрице. Если строка/столбец выражается как линейная комбинация других строк/столбцов, то информацию в строке/столбце можно считать избыточной. Соответственно, ранг матрицы снижается на 1.

Определитель, близкий к нулю, указывает на наличие мультиколлинеарности в данных.

## <center>ЧТО ТАКОЕ ФУНКЦИЯ?

В математике **функция** — это соответствие между двумя множествами, где каждому элементу из одного множества (области определения, *domain of a function*) сопоставляется единственный элемент из другого множества (области значений, *range of a function*).

Функция обозначается символом  и записывается в виде $f(x)$, где $x$ — элемент из области определения функции.

**Область определения** функции — это множество всех значений аргумента (переменной $x$).

**Область значений** — это множество всех значений функции (переменной $y$).

Область определений и область значений функции могут быть любыми множествами, в том числе множествами действительных чисел, комплексных чисел, векторов и т. д. Обычно функции в математическом анализе отображают элементы из множества вещественных чисел на другие вещественные числа.

Функции описывают зависимости и отношения между различными величинами: например, зависимость времени от скорости, температуры от высоты. Также можно рассчитать значение функции в определённой точке.

Функции можно задать разными способами — аналитически, графически или таблично.

* **Аналитически заданная функция представляет собой формулу**, которая позволяет вычислить значение функции для любого заданного значения аргумента.

* **Графически заданная функция представляет собой кривую**, которая отображает значения функции на оси координат.

* **Таблично заданная функция представляет собой список пар значений**, где первый элемент пары является аргументом функции, а второй — её значением.

В математическом анализе с помощью функций решают задачи оптимизации и дифференциальные уравнения, находят экстремумы и т. д. Кроме того, функции — основа для изучения других математических дисциплин, например теории вероятностей и статистики.

## <center>СВОЙСТВА ФУНКЦИЙ

*В математике у функций есть множество свойств, некоторые из которых особенно важны для их понимания и использования.*

Ниже перечислены некоторые из основных свойств функций:

* **Однозначность**. Каждому элементу из области определения функции соответствует единственный элемент из её области значений.

* **Обратимость**. Функция называется обратимой, если каждому элементу из области значений функции соответствует единственный элемент из её области определения.

* **Непрерывность**. Функция называется непрерывной, если изменение значения функции в результате бесконечно малого изменения аргумента также является бесконечно малым.

* **Гладкость**. Функция называется гладкой, если её производные любого порядка непрерывны.

* **Ограниченность**. Функция называется ограниченной, если её значения на области определения ограничены константой.

* **Монотонность**. Функция называется монотонной, если она убывает или возрастает на всей области определения или на некотором её интервале.

* **Периодичность**. Функция называется периодической, если её значения повторяются через определённый интервал.

$$\large f(x)=ax+b$$
Свойства функции $y = ax + b$:

**Гладкость.** Функция $y = ax + b$ гладкая на всей области определения, т. е. не имеет разрывов, углов и изломов.

**Ограниченность.** Функция $y = ax + b$ не является ограниченной сверху или снизу на всей области определения, за исключением случая, когда $a=0$. Тогда функция будет ограничена только снизу, если $b<0$, или ограничена только сверху, если $b>0$.

**Монотонность.** Функция $y = ax + b$ может быть монотонной или не монотонной в зависимости от знака коэффициента .
* Если $a>0$, функция возрастает.
* Если $a<0$, функция убывает.
* Если $a=0$, то функция — константа и не является монотонной.

**Периодичность**. Функция $y = ax + b$ не является периодической на всей области определения, за исключением случая, когда $a=0$. Тогда функция является периодической с любым периодом, проходящим через начало координат.

Свойства функции $f(x)=ax^2+b$:

* **Гладкость.** Функция $f(x)=ax^2+b$ гладкая на всей области определения, т. е. имеет непрерывные производные любого порядка.

* **Ограниченность.** Функция $f(x)=ax^2+b$ может быть ограниченной сверху или снизу на всей области определения в зависимости от знака коэффициента $a$ и значения $b$.

## <center>ЧТО ТАКОЕ ПРОИЗВОДНАЯ?

**Производная** — это понятие, которое используется в математике для описания того, как быстро изменяется функция в каждой точке. Если мы говорим о функции, которая описывает зависимость одной переменной от другой, то производная показывает, как быстро изменяется значение функции с изменением её аргумента в каждой точке.

Более формально производная функции определяется как предел отношения приращения функции к приращению её аргумента при стремлении приращения аргумента к нулю:

$$f^{\prime}(x)=\frac{f(x+\Delta x)-f(x)}{\Delta x}$$

, где $\Delta x$ — бесконечно малое приращение аргумента.

Эта формула показывает, как быстро изменяется функция $f(x)$ в точке $x$:

* Если производная положительная, функция возрастает в этой точке.

* Если производная отрицательная, функция убывает в этой точке.

* Если производная равна нулю, функция имеет экстремум (максимум или минимум) в этой точке.

*Производная — ключевой инструмент для решения многих задач, так как описывает изменение величин в различных процессах.*

Существует несколько способов записи производной функции. Вот наиболее распространённые:

* Производная в нотации Лейбница (обычная форма): $\frac{dy}{dx}$.

  Здесь $y$ — это функция от $x$, а дробь указывает на производную функции $y$ по переменной $x$. Если нужно указать производную более высокого порядка, то можно использовать запись $\frac{d^ny}{dx^n}$. Здесь $n$ указывает на порядок производной.

* Производная в нотации Лагранжа: $f^{(n)}\left(x_0\right)=\frac{n !}{2 \pi i} \oint_\gamma \frac{f(z)}{\left(z-x_0\right)^{n+1}} d z$.

  Здесь $f^{(n)}\left(x_0\right)$ означает $n$-ю производную функции $f(x)$ в точке $x_0$, $\gamma$ — контур, ограничивающий область, содержащую точку $x_0$.

* Производная в инфиксной нотации: $y^{\prime}(x)$ или $f^{\prime}(x)$. Эта запись эквивалентна нотации Лейбница $\frac{dy}{dx}$.

* Дифференциал функции: $df=f^{\prime}(x)dx$. Эта запись означает дифференциал функции $f(x)$, который можно представить как произведение производной $f^{\prime}(x)$ и бесконечно малого приращения $dx$.

Выбор записи производной зависит от предпочтений и требований к форматированию в конкретном случае.

* Если производная положительная, то наклон касательной положительный, что означает, что функция в этой точке возрастает.

* Если производная отрицательная, то наклон касательной отрицательный, что означает, что функция в этой точке убывает.

* Если производная равна нулю, то касательная к графику функции является горизонтальной и функция имеет экстремум (максимум или минимум) в этой точке.

Если график функции $y=f(x)$ имеет касательную в точке $(a, f(a))$, то производная функции в этой точке равна тангенсу угла наклона касательной к графику в этой точке. Пусть $m$ — угол наклона касательной к графику функции $y=f(x)$ в точке $(a, f(a))$. Тогда, если мы рассмотрим малый прирост аргумента $\Delta x$ и соответствующий прирост функции $\Delta y$, то получим следующее:
$$\operatorname{tan(m)}=\lim _{\Delta x \rightarrow 0} \frac{\Delta y}{\Delta x}$$
Если существует производная функции $f(x)$ в точке $a$, то этот предел существует и равен значению производной $f^{\prime}(a)$, т. е.
$$\operatorname{tan(m)}=f^{\prime}(a)$$
Таким образом, производная функции в точке касания графика касательной к этому графику равна тангенсу угла наклона касательной в этой точке.

Соответственно, функция является недифференцируемой, если:

* имеются углы;
* разрывы;
* точки с вертикальными асимптотами тангенса.

Недифференцируемые функции могут вызывать проблемы при решении математических задач. Вот некоторые из таких проблем:

* **Ограничение области определения.** Некоторые задачи требуют дифференцируемых функций, и использование недифференцируемых может ограничить область определения их применения.

* **Сложность вычислений.** Недифференцируемые функции могут быть более сложными в вычислении, так как их производные могут иметь особенности, например разрывы или бесконечные значения.

* **Несуществование градиента.** Некоторые задачи оптимизации требуют нахождения градиента функции, а градиент в недифференцируемых функциях не существует в некоторых точках.

* **Необходимость использования других методов.** В некоторых задачах требуется проанализировать недифференцируемые функции, и для их решения могут понадобиться другие методы, например методы интегрального и функционального анализа.

## <center>ПРАВИЛА ДИФФЕРЕНЦИРОВАНИЯ

Основные правила дифференцирования:

* Сумма производных:
$$(f(x)+g(x))^{\prime}=f^{\prime}(x)+g^{\prime}(x)$$
* Разность производных:
$$(f(x)-g(x))^{\prime}=f^{\prime}(x)-g^{\prime}(x)$$
* Произведение производных:
$$(f(x)g(x))^{\prime}=f^{\prime}(x)g(x)+f(x)g^{\prime}(x)$$
* Частное производных:
$$\left(\frac{f(x)}{g(x)}\right)^{\prime}=\frac{f^{\prime}(x) g(x)-f(x) g^{\prime}(x)}{[g(x)]^2}$$
* Производная произведения функции на константу равна произведению константы и производной функции:
$$(cf(x)){\prime}=cf{\prime}(x)$$
* Производная константы равна нулю:
$$\frac{d}{dx}[f(x)+C]=f^{\prime}(x)$$
* Правило производной обратной функции — производная обратной функции равна обратной производной функции в точке обратимости:
$$\left(f^{-1}(y)\right)^{\prime}=\frac{1}{f^{\prime}\left(f^{-1}(y)\right)}$$
* Правило цепочки — производная композиции функций равна произведению производной внешней функции и производной внутренней функции:
$$(f(g(x)))^{\prime}=f^{\prime}(g(x))g^{\prime}(x)$$
  Это правило играет важную роль в машинном обучении. В нотации Лейбница его можно записать так:
$$\frac{df}{dx}=\frac{df}{dg}\cdot \frac{dg}{dx}$$
Эта формула показывает, что производная сложной функции $f(g(x))$ равна произведению производной внешней функции $f^{\prime}(g(x))$ и производной внутренней функции $g^{\prime}(x)$. Например, если $f(x)=sin(x^2)$ и $g(x)=x^2$, то
$$\frac{d f}{d x}=\frac{d}{d x} \sin (g(x))=\cos (g(x)) \cdot \frac{d g}{d x}=\cos \left(x^2\right) \cdot 2 x$$
Это правило можно использовать для дифференцирования более сложных функций, которые представляют собой композицию нескольких функций.

*Эти правила можно расширить и использовать в более сложных случаях, но они образуют основу дифференцирования функций.*

## <center>ПРОИЗВОДНЫЕ ЧАСТО ИСПОЛЬЗУЕМЫХ ФУНКЦИЙ

* Константа:
$$f^{\prime}(C)=0$$
* Линейная функция:
$$f^{\prime}(C)=0$$
* Степенная функция:
$$f^{\prime}(x^n)=nx^{n-1}$$
* Степенная функция с отрицательным целым показателем:
$$f^{\prime}(x^{-n})=-nx^{-n-1}$$
* Тригонометрические функции:
$$f^{\prime}(\sin(x))=\cos(x)$$
$$f^{\prime}(\sin(x))=\cos(x)$$
$$f^{\prime}(\tan(x))=\sec^2(x)$$
* Обратные тригонометрические функции:
$$f^{\prime}(\arcsin (x))=\frac{1}{\sqrt{1-x^2}}$$
$$f^{\prime}(\arccos (x))=-\frac{1}{\sqrt{1-x^2}}$$
$$f^{\prime}(\arctan (x))=\frac{1}{1+x^2}$$
* Экспоненциальная функция:
$$f^{\prime}(e^x)=e^x$$
* Логарифмическая функция:
$$f^{\prime}(\ln(x))=\frac{1}{x}$$

*Эти производные можно использовать для нахождения производных более сложных функций, которые представляют собой композицию нескольких функций.*

## <center>РЕАЛИЗАЦИЯ ПРОИЗВОДНЫХ В PYTHON

В этом юните мы поговорили о производных. Давайте вкратце повторим основные моменты.

**Производная функции** — это показатель скорости изменения значения функции в каждой её точке. Математически производная функции $f(x)$ в точке $x_0$ определяется как предел отношения изменения значения функции $f(x)$ к изменению её аргумента $x$ при стремлении изменения аргумента к нулю:
$$f^{\prime}(x)=\frac{f(x+\Delta x)-f(x)}{\Delta x}$$
, где $\Delta x$ — бесконечно малое приращение аргумента.

Производные функций широко используются в машинном обучении, например для оптимизации функций потерь при обучении нейронных сетей. Оптимизация функций потерь требует нахождения экстремумов функции. Найти их можно, вычислив производные и используя методы оптимизации, такие как градиентный спуск. Производные также можно использовать для нахождения экстремумов функций при обучении моделей, таких как линейная и логистическая регрессия.
