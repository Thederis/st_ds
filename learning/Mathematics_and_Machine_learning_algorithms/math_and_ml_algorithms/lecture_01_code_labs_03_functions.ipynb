# Построение графиков  <a class="anchor" id="plotting"></a>

Matplotlib - это библиотека на языке программирования Python для создания графиков и визуализации данных. Она предоставляет широкий спектр возможностей для создания различных видов графиков, включая:

- Линейные графики: простые, многострочные, точечные, шаговые, шестиугольные и т.д.
- Гистограммы: вертикальные, горизонтальные, наклонные, с различными стилями заполнения и т.д.
- Круговые диаграммы: стандартные, вложенные, секторные и т.д.
- Графики рассеяния: одномерные, двумерные, многомерные, с настройкой размера и цвета точек и т.д.
- Контурные и поверхностные графики: линии уровня, изолинии, тепловые карты, трехмерные графики и т.д.
- Диаграммы рассеивания и ящики с усами: создание диаграмм рассеивания, ящиков с усами, настраиваемых маркеров и цветов.
- Графики временных рядов: создание линейных графиков временных рядов с настраиваемыми осями и метками времени.
- Диаграммы ошибок: графики ошибок, погрешностей и отклонений, графики с асимметричными ошибками и т.д.
- Анимации: создание анимированных графиков, включая изменение позиции, размера, цвета и т.д.

Это далеко не полный список возможностей, которые предоставляет библиотека Matplotlib. Она также имеет широкий спектр настроек и параметров для настройки внешнего вида графиков и диаграмм, таких как цвета, шрифты, легенды, подписи и т.д.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## f(x) = ax + b <a class="anchor" id="linear"></a>

Для построения графика функции $y = a*x + b$ в библиотеке Matplotlib необходимо выполнить следующие шаги:

- Импортировать библиотеку Matplotlib и библиотеку NumPy для работы с массивами данных.
- Определить значения x, которые будут использоваться для построения графика. Например, можно создать массив значений от -10 до 10 с помощью функции numpy.linspace.
- Вычислить значения y для каждого значения x, используя формулу y = a*x + b.
- Использовать функцию matplotlib.pyplot.plot для построения графика, передав значения x и y в качестве аргументов. Дополнительно можно настроить цвет и стиль линии с помощью дополнительных аргументов функции.

```python
# Определение значений x
x = np.linspace(-10, 10, 100)

# Определение параметров a и b
a = 2
b = 1

# Вычисление значений y
y = a*x + b

# Построение графика
plt.plot(x, y, color='blue', linestyle='-', linewidth=2)

# Настройка осей координат и заголовка
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции y = a*x + b')
plt.axis('equal')
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Отображение графика
plt.show()
```

## f(x) = $x^2$ <a class="anchor" id="parabola"></a>

Для построения графика функции y = $x^2$ в библиотеке Matplotlib необходимо выполнить следующие шаги:

- Импортировать библиотеку Matplotlib и библиотеку NumPy для работы с массивами данных.
- Определить значения x, которые будут использоваться для построения графика. Например, можно создать массив значений от -10 до 10 с помощью функции numpy.linspace.
- Вычислить значения y для каждого значения x, используя формулу y = $x^2$.
- Использовать функцию matplotlib.pyplot.plot для построения графика, передав значения x и y в качестве аргументов. Дополнительно можно настроить цвет и стиль линии с помощью дополнительных аргументов функции.

```python
# Определение значений x
x = np.linspace(-10, 10, 100)

# Вычисление значений y
y = x**2

# Построение графика
plt.plot(x, y, color='blue', linestyle='-', linewidth=2)

# Настройка осей координат и заголовка
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции y = x**2')
plt.axis('equal')
plt.xlim(-10, 10)
plt.ylim(-1, 10)

# Отображение графика
plt.show()
```

## f(x) = $\frac{1}{x}$ <a class="anchor" id="hyperbola"></a>

Для построения графика функции y = $\frac{1}{x}$ в библиотеке Matplotlib необходимо выполнить следующие шаги:

- Импортировать библиотеку Matplotlib и библиотеку NumPy для работы с массивами данных.
- Определить значения x, которые будут использоваться для построения графика. Например, можно создать массив значений от -10 до 10 с помощью функции numpy.linspace, за исключением точки x = 0, так как в этой точке функция не определена.
- Вычислить значения y для каждого значения x, используя формулу y = 1/x, за исключением точки x = 0.
- Использовать функцию matplotlib.pyplot.plot для построения графика, передав значения x и y в качестве аргументов. В данном случае придется строить 2 графика, чтобы отобразить разрывность в точке 0. Дополнительно можно настроить цвет и стиль линии с помощью дополнительных аргументов функции.

```python
# Определение значений x, за исключением точки x=0
x = np.linspace(-10, 10, 200)
x1 = x[:100]
x2 = x[101:]

# Вычисление значений y, за исключением точки x=0
y1 = 1/x1
y2 = 1/x2

# Построение графика
plt.plot(x1, y1, color='blue', linestyle='-', linewidth=2)
plt.plot(x2, y2, color='blue', linestyle='-', linewidth=2)

# Настройка осей координат и заголовка
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции y = 1/x')

# Устанавливаем пределы оси x
plt.axis('equal')
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Отображение графика
plt.show()
```
