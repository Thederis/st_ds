### Пример
**Условие.** Напишите программу, которая просит пользователя ввести электронную почту и проверяет корректность её введения. По условию, почтовый адрес является корректным, если в нём присутствуют один и только один символ `@` (при этом слева от этого символа должен находиться хотя бы ещё один [исключая `@`]). _Это не полный список критериев, но для учебного примера достаточно этих ограничений._ При нарушении хотя бы одного условия программа должна ответить, что такая почта не принимается, и разъяснить причину. Если почта корректна, об этом также надо сообщить пользователю.

**Алгоритм решения:**
1. Получить на вход почту в виде строки (функция `input`).
2. Проверить разные условия (условная конструкция и подходящие операторы):
 - проверить наличие ровно одного символа `@` (строковый метод `.count()`) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`);
 - проверить наличие символов перед `@` (строковый метод `.startswith()` — если начинается с `@`, то условие не выполнено) и вывести отказ с пояснением, если проверка не пройдена (функция `print()`).
3. Сообщить пользователю, что почта корректна, если это так (функция `print()`).

**Решение:**

````python
email = input()  # шаг 1

if email.count('@') != 1:  # шаг 2-1
    print('Введённый почтовый адрес некорректен. Он должен содержать ровно одну @')
elif email.startswith('@'):  # шаг 2-2
    print('Введённый почтовый адрес некорректен. Он должен иметь хотя бы один символ перед @')
else:  # шаг 3
    print('Поздравляем, ваша почта принята!')
````
