Напишите функцию holes_count(), которая принимает на вход число number, подсчитывает общее количество отверстий в заданном числе и возвращает результат.

Условимся, что цифрами с отверстиями будем считать 0, 4, 6, 8 и 9. Причём в цифре 8 отверстий два, а в остальных — по одному.

Обратите внимание, что функция должна возвращать суммарное количество отверстий по всем цифрам в числе.

```python
def holes_count(number):
    count = 0
    for i in str(number):
        if i in ['0', "4", "6", "9"]:
            count += 1
        if i == "8":
            count += 2
    return count




print(holes_count(8))
# 2
print(holes_count(146))
# 2
print(holes_count(84628))
# 6
```

```python
s = "ddvvrwwwrggg"

j = 1
n = 0
for i in range(len(s)-1):
  if s[i+1] == s[i]:
    j += 1
  elif s[i+1] != s[i]:
    j = 1
  if j > n:
    n = j

print(n)

#Например:

# "" -> 0
# "a" -> 1
# "abbc" -> 2
# "adddaabaa" -> 3
```

Напишите функцию sum_min_numbers(), которая также принимает на вход три числа (a, b, c) и возвращает сумму двух наименьших.

Можно использовать функцию find_min_number() из предыдущего задания для поиска минимального числа.

Примеры работы программы:


print(sum_min_numbers(1, 2, 3))

 3

print(sum_min_numbers(1, 2, -10))

 -9

```python
def sum_min_numbers(a, b, c):

    def find_min_number1(a, b, c):
        return min([a, b, c])

    numbers = [a, b, c]
    numbers.remove(find_min_number1(a, b, c))

    def find_min_number2(*args):
        return min(*args)

    return find_min_number1(a, b, c) + find_min_number2(numbers)


print(sum_min_numbers(1, 2, 3))
# 3
print(sum_min_numbers(1, 2, -10))
# -9
```

Примечание. Билет счастливый, если сумма первых трёх цифр в его номере равна сумме последних трёх цифр.

На вход функция принимает шестизначное число ticket_number и должна возвращать одно из булевых значений (True или False) в зависимости от того, является ли билет счастливым.

При решении постарайтесь не использовать встроенную функцию sum() — примените циклы.

Примеры работы программы:


print(lucky_ticket(111111))

 -  True

print(lucky_ticket(123456))

 -  False

```python
def lucky_ticket(number):
    number_str = str(number)
    count_l = 0
    for i in range(3):
        count_l += int(number_str[i])
    count_r = 0
    for i in range(1, 4):
        count_r += int(number_str[-i])
    return count_l == count_r

print(lucky_ticket(111111))
# True
print(lucky_ticket(123456))
# False
```
