<h1>Иерархия потребностей</h1>
<p>Функции, которые выполняют дата-инженер и дата-сайентист, а также классификация этих функций хорошо представлены на рисунке ниже. Пирамида, изображающая иерархию нужд в Data science, наталкивает на мысль, что верхние слои невозможны без нижних. Забегая вперёд, заметим, что Data Engineering тяготеет к нижней части этой пирамиды.</p>

<div class="h1">Эволюционный взгляд</div>
<p>С другой стороны, обозначить границы профессий Data Engineer (DE) и Data Scientist (DS) позволяет эволюционный взгляд. Рассмотрим пошагово:</p>
<ul class="list">
<li>В компанию N приходит DS, его задача — на основе имеющихся данных построить модель машинного обучения для расчёта стоимости квартир.</li>
<li>Вскоре DS понимает, что 90 % его рабочего времени занимает не построение модели, а иные задачи. После анализа ситуации оказывается, что большую часть времени DS подготавливает данные.</li>
<li>DS хочет делегировать задачи по подготовке данных новому специалисту — DE.</li>
</ul>
<p>Так в отдельно взятом случае рождается инженерия данных.</p>
<h1>Синергия DS и DE</h1>
<ul class="list">
<li>Роли<strong> </strong>DS и DE очень синергичны. Разделение обязанностей между ними условно и зависит от персональных особенностей и возможностей конкретных специалистов.</li>
<li>Несмотря на то что оба специалиста нужны, в маленьких компаниях их роли часто объединены, как в нашем примере, где при построении модели расчёта стоимости квартир дата-сайентист выполняет и функции дата-инженера.</li>
<li>Если в компании есть и DE, то он выполняет вспомогательную роль, беря на себя задачи по подготовке данных для DS.</li>
<li>Но в то же время, как ни парадоксально, DE может существовать и без DS. В этом случае результатами его труда пользуются BI-аналитики (Business Intelligence), которым тоже нужны подготовленные данные.</li>
</ul>
<h1>Что делает DE</h1>
<p>Вернёмся к пирамиде с иерархией потребностей в Data Science и перечислим функции дата-инженера.</p>
<ul class="list">
<li>В<strong> </strong>Move/Store (хранение и перемещение данных) это:
<ul class="list">
<li>хранение структурированных и неструктурированных данных;</li>
<li>построение и эксплуатация конвейеров и ETL (от англ. "Extract, Transform, Load");</li>
<li>управление всей инфраструктурой данных;</li>
<li>обеспечение надёжных потоков данных.</li>
</ul>
</li>
<li>В Explore/Transform (преобразования) это:
<ul class="list">
<li>очистка данных;</li>
<li>работа с аномалиями;</li>
<li>подготовка данных.</li>
</ul>
</li>
</ul>
<h2>Основные инструменты DE:</h2>
<ul class="list">
<li>Python — язык (желательны более «быстрые» языки Scala/Java/C++);</li>
<li>Jupyter Notebook — основной рабочий инструмент (DE и DS);</li>
<li>Spark — основной фреймворк для работы с данными;</li>
<li>HDFS, Hive — основное хранилище;</li>
<li>JDBC, SQL — основные источники (структурированные данные);</li>
<li>файлы, Kafka — источники (неструктурированные данные);</li>
<li>Airflow — оркестрация конвейеров обработки данных.</li>
</ul>

#  3. Что такое базы данных?

<div class="main-block">
<h1>Что такое базы данных?</h1>
<p style="font-size: 16px;">В современном мире базы данных используются практически везде. Например, когда вы заходите на сайт и вводите логин и пароль, программа проверяет, есть ли такая пара в базе данных. В базах хранятся данные о посещениях, покупках, контактах пользователей (или покупателей). Если данных становится много, на помощь приходят <strong>базы данных (БД).</strong></p>
<div class="term"><strong>База данных</strong> — это организованная структура для хранения, изменения и обработки взаимосвязанной информации. Обычно базы данных используют для взаимодействия с большими объёмами данных. Практически ни одна большая компания (от интернет-магазинов до банков) сейчас не обходится без использования баз данных.</div>

<div class="term"><strong>В реляционных БД таблица</strong> — это набор элементов данных (значений), использующий модель вертикальных столбцов (имеющих уникальное имя) и горизонтальных строк.</div>
<div class="term"><strong>Ячейка</strong> — место, где строка и столбец пересекаются. Таблица содержит определённое число столбцов, но может иметь любое количество строк.</div>

<div class="term">Иными словами, <strong>первичный ключ</strong> — это подмножество столбцов, которое уникально идентифицирует строку. Часто в качестве первичного ключа берут просто ID объекта, например, ID пользователя или ID покупки.</div>

# 9. Типы данных в SQL

<div class="main-block">
<p>В основном используются следующие типы данных:</p>
<ul class="list">
<li><code style="background-color: #74992e;">integer</code> (или <code style="background-color: #74992e;">bigint</code>) — применяется для больших чисел.</li>
<li><code style="background-color: #74992e;">numeric</code> — особенно рекомендуется для хранения денежных сумм и других величин, где важна точность.</li>
<li><code style="background-color: #74992e;">serial</code> — используется для создания первичных ключей таблицы (обычно для создания ID строки). Он очень удобен, так как при добавлении строк в таблицу сам проставляет им уникальный ID на 1 больше, чем предыдущий проставленный ID.</li>
<li><code style="background-color: #74992e;">real</code> — подходит для чисел, где хорошая точность не очень нужна.</li>
